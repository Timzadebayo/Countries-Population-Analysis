```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_population_raw = pd.read_csv('population_total.csv')
```

```python
df_population_raw
```

# using .pivot

```python
# dropping null values  in the data set we use .dropna()
```

```python
df_population_raw.dropna()   # use shift + Tab to know the arugument in a function
```

```python
# To effect the changes of the .dropna() function we use inplace function or we overwrite it
```

```python
df_population_raw = df_population_raw.dropna()   #to redefine the data set
```

```python
df_population_raw
```

```python
df_population_raw.pivot(index = 'year' , columns = 'country' , values = 'population')   #reshape helps to re-arrange it
```

```python
reshaped_dataframes = df_population_raw.pivot(index = 'year' , columns = 'country' , values = 'population')
```

```python
reshaped_dataframes
```

```python
reshaped_dataframes[['United States' , 'China']] #if it is one it will be .unites states
```

```python
reshaped_dataframes[['United States' , 'China']]
```

```python
reshaped_dataframes[['Mali' , 'China']]
```

```python
selected_country=reshaped_dataframes[['China','Nigeria','Gambia', 'Argentina','Ghana']]
```

```python
selected_country
```

```python
# To use the line plot, we use the .plot function
```

```python
selected_country.plot(kind = 'line', xlabel = 'Year' , ylabel = 'population', title = 'Line Graph' , figsize = (8,8))
plt.savefig('pivot.png')
plt.show()


```

```python
selected_country.to_excel('Pivot 2 Table.xlsx')
```
